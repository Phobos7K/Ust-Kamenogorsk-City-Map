```python
import matplotlib.pyplot as plt
import osmnx as ox
```

```python
#getting the boundary polygon for Oskemen, project it, and plot it!
area = ox.geocode_to_gdf("Ust-Kamenogorsk")
area_polygon = area.geometry.values[0]
area_polygon
#ax = area_polygon.plot(fc = "gray", ec = "none")
#ax.axis("off")
```

Getting the area snapshot from polygon

```python
area_polygon = ox.graph_from_polygon(area_polygon)
area_polygon.number_of_nodes(), area_polygon.number_of_edges()
```

```python
nodes, edges = ox.graph_to_gdfs(area_polygon)
plt.style.use("ggplot")
```

```python
f, ax = plt.subplots(1, figsize = (10, 10))
edges.plot(ax = ax, edgecolor = "green", linewidth = 0.3)
ax.set_title("Ust-Kamenogorsk", fontsize = 20, color = "black", fontname = "Sugoe UI")
ax.axis("off")
```
